```python
# Extract indices when variables satisfy certain conditions
# lev for geopotential height

ds_nonrr = xr.open_dataset("/home/phymku/Data/RR_ANALYSIS/datasets/SOR_NONRR_DATA/FWmaHIST_ne30pg3_ne30pg3_mg17_no_conus_gravity_waves_mmk_subset2")

da = ds_nonrr['mmk_z3'].isel(ncol=0).isel(time=0) # Isolate data - geopotential height

errtol = 1 # choose error tolerance

data = da.values # extract numerical values from dataArray
a = np.argwhere(np.abs(data - 80) < errtol).tolist() # Return indices where condition in np.argwhere() is satisfied
print(a)
# put all indices in a list that can be used to index xr dataArray
b = [i[0] for i in a] # The i[0] is there because without it you get a list of lists, where each individual index is its own "list"
print(b)
da.isel(lev=b)
```

```python
# Even easier
desired_value = 2
a = np.abs(da.values - desired_value).argmin()
print("Index of desired value ", a)
print("Desired_value ",  desired_value)
print("Returned value ", da.isel(lev=a)['mmk_z3'].values)
```
