```python
# Structure of plotting script
# Plot controls
# Plot Non-RR
# Non-RR contours
# Plot RR
# RR contours
# diffs
# diff contours

fig, ax = plt.subplots(nrows=1,ncols=3,figsize=(25,7.5))

ylimL = 3e3
ylimU = 1e-6
vmin_nrr = 150
vmax_nrr = 600
vmin_rr = vmin_nrr
vmax_rr = vmax_nrr
vmin_diff = -100
vmax_diff = 50
contours = 1 # boolean
contfont = 10

ds_nonrr_var.plot(ax=ax[0],y="lev",yscale="log",ylim=[ylimL,ylimU],cmap="turbo",vmin=vmin_nrr,vmax=vmax_nrr)
if contours:
    # lined contours
    cs_nonrr = ds_nonrr_var.plot.contour(y="lev",
                                         yscale="log",
                                         vmin=vmin_nrr,vmax=vmax_nrr,
                                         ylim=[ylimL,ylimU],
                                         levels=10,
                                         colors="k",
                                         linewidths=0.5,
                                         ax=ax[0],
                                         # norm=log_norm, # uncomment for log scale - not working properly for contours!
                                         )

    # color label controls
    fmt = {}
    for l in cs_nonrr.levels:
        # l = l # uncomment for log
        fmt[l] =f"{l:.2f}" # uncomment for nolog
        # fmt[l] =f"{l:.3f}" # uncomment for log
    clabels = ax[0].clabel(cs_nonrr, cs_nonrr.levels, inline=True, fmt=fmt, fontsize=contfont)
    [txt.set_bbox(dict(facecolor='white', edgecolor='none', pad=0)) for txt in clabels]

    
ds_rr_var.plot(ax=ax[1],y="lev",yscale="log",ylim=[ylimL,ylimU],cmap="turbo",vmin=vmin_rr,vmax=vmax_rr)
if contours:
    # lined contours
    cs_rr = ds_rr_var.plot.contour(y="lev",
                                   yscale="log",
                                   vmin=vmin_rr,vmax=vmax_rr,
                                   ylim=[ylimL,ylimU],
                                   levels=10,
                                   colors="k",
                                   linewidths=0.5,
                                   ax=ax[1],
                                    # norm=log_norm, # uncomment for log scale - not working properly for contours!
                                   )

    # color label controls
    fmt = {}
    for l in cs_rr.levels:
        # l = l # uncomment for log
        fmt[l] =f"{l:.2f}" # uncomment for nolog
        # fmt[l] =f"{l:.3f}" # uncomment for log
    clabels = ax[1].clabel(cs_rr, cs_rr.levels, inline=True, fmt=fmt, fontsize=contfont)
    [txt.set_bbox(dict(facecolor='white', edgecolor='none', pad=0)) for txt in clabels]

diff_var.plot(ax=ax[2],y="lev",yscale="log",ylim=[ylimL,ylimU],cmap="turbo",vmin=vmin_diff,vmax=vmax_diff)
diff_type = "RR - Non-RR absolute"
if contours:
    # lined contours
    cs_diff = diff_var.plot.contour(y="lev",
                                    yscale="log",
                                    vmin=vmin_diff,vmax=vmax_diff,
                                    ylim=[ylimL,ylimU],
                                    levels=10,
                                    colors="k",
                                    linewidths=0.5,
                                    ax=ax[2],
                                    # norm=log_norm, # uncomment for log scale - not working properly for contours!
                                       )

    # color label controls
    fmt = {}
    for l in cs_diff.levels:
        # l = l # uncomment for log
        fmt[l] =f"{l:.2f}" # uncomment for nolog
        # fmt[l] =f"{l:.3f}" # uncomment for log
    clabels = ax[2].clabel(cs_diff, cs_diff.levels, inline=True, fmt=fmt, fontsize=contfont)
    [txt.set_bbox(dict(facecolor='white', edgecolor='none', pad=0)) for txt in clabels]


ax[0].set_ylabel("Pressure (hPa)")
ax[1].set_ylabel("Pressure (hPa)")
ax[2].set_ylabel("Pressure (hPa)")
ax[0].set_title(var + " Non-RR")
ax[1].set_title(var + " RR")
ax[2].set_title(var + " diffs " + diff_type)
```
